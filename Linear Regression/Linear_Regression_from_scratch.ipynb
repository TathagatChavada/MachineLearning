```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Set seed for reproducibility
np.random.seed(42)

# Generate random integer data for x and y
x = np.random.randint(20, 91, size=100)  # 100 data points between 20 and 90
y = 3 * x + np.random.randint(-10, 11, size=100)  # Assuming a linear relationship with some random noise

plt.scatter(x, y, label='Generated Data')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Regression Data Generation')

plt.legend()

plt.show()
```

```python
def gradient_descent(current_m, current_b, x, y, learning_rate):

    m_gradient = 0
    b_gradient = 0

    n = len(x)

    for i in range(n):
        m_gradient += -(2/n) * x[i] * (y[i] - (current_m * x[i] + current_b))
        b_gradient += -(2/n) * (y[i] - (current_m * x[i] + current_b))


    m = current_m - m_gradient * learning_rate
    b = current_b - b_gradient * learning_rate

    return m, b

```

```python
m, b = 0,0

learning_rate = 0.0001
epochs = 1000


for i in range(epochs):

    if (i % 100 == 0):
        print(f"Epoch: {i}")

    m, b = gradient_descent(m, b, x, y, learning_rate)


print(f"Slope: {m} , Interscept: {b}")

plt.scatter(x, y, label='Generated Data')
plt.plot(list(range(20, 90)), [m * x + b for x in range(20,90)], color="red")
plt.show()
```
